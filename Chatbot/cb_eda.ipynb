```python
import pandas as pd     
import numpy as np
```

```python
train= pd.read_excel("데이터/Training.xlsx")
val=pd.read_excel('데이터/Validation.xlsx')
```

```python
df=pd.concat([train,val],ignore_index=True)
df.head()
```

데이터 합치기 

```python
# 질문과 답변만 추출 
df1= df[['사람문장1','시스템문장1']]
df1.columns=['question', 'answer']
df2= df[['사람문장2','시스템문장2']]
df2.columns=['question', 'answer']
df3= df[['사람문장3','시스템문장3']]
df3.columns=['question', 'answer']
```

```python
# 각 추출한 질문과 답변 합치기 
df4=pd.concat([df1,df2,df3],ignore_index=True)
df4
```

```python
#결측값 확인
df4.isnull().sum()
```

```python
#결측값 제거 
df4= df4.dropna(axis=0)
df4.head()
```
