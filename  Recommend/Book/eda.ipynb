Dataset Info.

train.csv [파일]
ID : 샘플 고유 ID
User-ID : 유저 고유 ID
Book-ID : 도서 고유 ID
유저 정보
Age : 나이
Location : 지역
도서 정보
Book-Title : 도서 명
Book-Author : 도서 저자
Year-Of-Publication : 도서 출판 년도 (-1일 경우 결측 혹은 알 수 없음)
Publisher : 출판사
Book-Rating : 유저가 도서에 부여한 평점 (0점 ~ 10점)
단, 0점인 경우에는 유저가 해당 도서에 관심이 없고 관련이 없는 경우
